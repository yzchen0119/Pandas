# PyCity Schools Analysis

* Top 5 schools are all charter schools with lower budgets, while bottom 5 schools are all district schools with higher budgets.

* The scores and passing rates of small and medium sized schools dramatically higher than large size schools(excpet the average reading scores).

* In a same school, the math and reading scores stay consistent across each grade. 

---

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load 
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```


## District Summary


```python
school_count = school_data_complete["school_name"].nunique()
student_count = school_data_complete["student_name"].count()
budget_count = school_data_complete["budget"].unique().sum()
avg_math_score = school_data_complete["math_score"].mean()
avg_read_score = school_data_complete["reading_score"].mean()
math_pass_df = school_data_complete[school_data_complete["math_score"]>=70]
percent_math_pass = (math_pass_df["student_name"].count()/student_count)*100
read_pass_df = school_data_complete[school_data_complete["reading_score"]>=70]
percent_read_pass = (read_pass_df["student_name"].count()/student_count)*100
percent_overall_pass = (percent_math_pass + percent_read_pass)/2

district_summary_df = pd.DataFrame({"Total Schools": [school_count], 
                                    "Total Students": [student_count], 
                                    "Total Budget": [budget_count],
                                    "Average Math Score": [avg_math_score], 
                                    "Average Reading Score": [avg_read_score],
                                    "% Passing Math": [percent_math_pass], 
                                    "% Passing Reading": [percent_read_pass], 
                                    "% Overall Passing Rate": [percent_overall_pass]})
district_summary_df.style.format({"Total Students": "{:,}", 
                                  "Total Budget": "${:,.2f}",
                                  "Average Math Score": "{:,.2f}", 
                                  "Average Reading Score": "{:,.2f}",
                                  "% Passing Math": "{:,.2f}", 
                                  "% Passing Reading": "{:,.2f}", 
                                  "% Overall Passing Rate": "{:,.2f}"})
```

## School Summary

```python
school_group_df = school_data_complete.set_index("school_name").groupby("school_name")
school_type = school_data.set_index("school_name")["type"]
school_student_count = school_group_df["student_name"].count()
school_budget = school_data.set_index("school_name")["budget"]
per_student_budget = school_budget/school_student_count
school_avg_math_score = school_group_df["math_score"].mean()
school_avg_read_score = school_group_df["reading_score"].mean()
school_percent_math_pass = math_pass_df.groupby("school_name")["student_name"].count()/school_student_count*100
school_percent_read_pass = read_pass_df.groupby("school_name")["student_name"].count()/school_student_count*100
school_overall_pass = (school_percent_math_pass + school_percent_read_pass)/2

school_sum_df = pd.DataFrame({"School Type": school_type, 
                              "Total Students": school_student_count, 
                              "Total School Budget": school_budget, 
                              "Per Student Budget": per_student_budget, 
                              "Average Math Score":school_avg_math_score, 
                              "Average Reading Score": school_avg_read_score, 
                              "% Passing Math":school_percent_math_pass, 
                              "% Passing Reading": school_percent_read_pass, 
                              "% Overall Passing Rate": school_overall_pass})

school_sum_df.head().style.format({"Total School Budget": "${:,.2f}", 
                            "Per Student Budget": "${:,.2f}", 
                            "Average Math Score":"{:,.2f}",
                            "Average Reading Score": "{:,.2f}", 
                            "% Passing Math":"{:,.2f}", 
                            "% Passing Reading": "{:,.2f}",
                            "% Overall Passing Rate": "{:,.2f}"})
```

## Top Performing Schools (By Passing Rate)

```python
top_school = school_sum_df.sort_values("% Overall Passing Rate", ascending = False)
top_school.head().style.format({"Total School Budget": "${:,.2f}", 
                                "Per Student Budget": "${:,.2f}", 
                                "Average Math Score":"{:,.2f}",
                                "Average Reading Score": "{:,.2f}", 
                                "% Passing Math":"{:,.2f}", 
                                "% Passing Reading": "{:,.2f}",
                                "% Overall Passing Rate": "{:,.2f}"})
```

## Bottom Performing Schools (By Passing Rate)

```python
bottom_school = school_sum_df.sort_values("% Overall Passing Rate", ascending = True)
bottom_school.head().style.format({"Total School Budget": "${:,.2f}", 
                                   "Per Student Budget": "${:,.2f}", 
                                   "Average Math Score":"{:,.2f}",
                                   "Average Reading Score": "{:,.2f}", 
                                   "% Passing Math":"{:,.2f}", 
                                   "% Passing Reading": "{:,.2f}",
                                   "% Overall Passing Rate": "{:,.2f}"})
```

## Math Scores by Grade

```python
grade_list = school_data_complete["grade"].unique()
grade_list
```

```python
grade9_data = school_data_complete[school_data_complete["grade"]=="9th"].groupby("school_name")
grade10_data = school_data_complete[school_data_complete["grade"]=="10th"].groupby("school_name")
grade11_data = school_data_complete[school_data_complete["grade"]=="11th"].groupby("school_name")
grade12_data = school_data_complete[school_data_complete["grade"]=="12th"].groupby("school_name")
avg_math_grade9 = round(grade9_data["math_score"].mean(), 2)
avg_math_grade10 = round(grade10_data["math_score"].mean(), 2)
avg_math_grade11 = round(grade11_data["math_score"].mean(), 2)
avg_math_grade12 = round(grade12_data["math_score"].mean(), 2)

grade_math_df = pd.DataFrame({"9th": avg_math_grade9, 
                              "10th": avg_math_grade10, 
                              "11th": avg_math_grade11, 
                              "12th": avg_math_grade12})
grade_math_df
```

## Reading Score by Grade 

```python
avg_read_grade9 = round(grade9_data["reading_score"].mean(), 2)
avg_read_grade10 = round(grade10_data["reading_score"].mean(), 2)
avg_read_grade11 = round(grade11_data["reading_score"].mean(), 2)
avg_read_grade12 = round(grade12_data["reading_score"].mean(), 2)

grade_read_df = pd.DataFrame({"9th": avg_read_grade9, 
                              "10th": avg_read_grade10, 
                              "11th": avg_read_grade11, 
                              "12th": avg_read_grade12})
grade_read_df
```

## Scores by School Spending

```python
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
school_sum_df["Spending Ranges (Per Student)"] = pd.cut(school_sum_df["Per Student Budget"], spending_bins, labels=group_names)
spending_df = school_sum_df.groupby("Spending Ranges (Per Student)")
spending_df = round(spending_df[["Average Math Score", 
                                 "Average Reading Score", 
                                 "% Passing Math", 
                                 "% Passing Reading", 
                                 "% Overall Passing Rate"]].mean(), 2)
spending_df
```

## Scores by School Size

```python
size_bins = [0, 1000, 2000, 5000]
group_names2 = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
school_sum_df["size"] = school_data.set_index("school_name")["size"]
school_sum_df["School Size"] = pd.cut(school_sum_df["size"], size_bins, labels=group_names2)
school_size_df = school_sum_df.groupby("School Size")
school_size_df = round(school_size_df[["Average Math Score", 
                                       "Average Reading Score", 
                                       "% Passing Math", 
                                       "% Passing Reading", 
                                       "% Overall Passing Rate"]].mean(), 2)
school_size_df
```
